```python
import pandas as pd
```

```python
df = pd.read_csv(
    "recipe_dataset.csv"
)

print(df.shape)

df.head()
```

```python
df.isnull().sum()
```

```python
#xóa các dòng trùng lặp dựa trên cột "url"
df.drop_duplicates(
    subset=["url"],
    inplace=True
)

print(df.shape)
```

```python
#xóa dl lỗi
df.dropna(
    subset=[
        "title",
        "ingredients",
        "instructions"
    ],
    inplace=True
)

print(df.shape)
```

```python
import re

TEXT_COLUMNS = [

    "title",

    "description",

    "ingredients",

    "instructions",

    "keywords"

]
# chỉ bỏ dấu "
df["title"] = (
    df["title"]
    .astype(str)
    .str.replace(
        '"',
        '',
        regex=False
    )
)
for col in TEXT_COLUMNS:

    if col in df.columns:

        TEXT_COLUMNS = [

            "title",

            "description",

            "ingredients",

            "instructions",

            "keywords"

        ]
        # chỉ bỏ dấu "
        df["title"] = (
            df["title"]
            .astype(str)
            .str.replace(
                '"',
                '',
                regex=False
            )
        )
        for col in TEXT_COLUMNS:

            if col in df.columns:

                df[col] = df[col].apply(
                    clean_text
                )
```

```python
# dùng cho ranking sau
df["popularity_score"] = (

    df["likes"] * 2

    +

    df["bookmarks"] * 3

    +

    df["comment_count"]

)
```

```python
# tính số lượng nguyên liệu
df["ingredient_count"] = (

    df["ingredients"]

    .str.split("|")

    .apply(len)

)
```

```python
#đếm số bước nấu ăn
df["cook_steps"] = (

    df["instructions"]

    .str.split("|")

    .apply(len)

)
```

```python
def remove_hashtags(text):

    if pd.isna(text):
        return ""

    text = re.sub(
        r"#\w+",
        "",
        str(text)
    )

    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text.strip()

df["description"] = (
    df["description"]
    .apply(remove_hashtags)
)
```

```python
df["recipe_text"] = (

    df["title"]

    + " "

    + df["description"]

    + " "

    + df["ingredients"]

)
```

```python
df["recipe_text"] = (
    df["recipe_text"]
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
```

```python
if "ingredients_list" in df.columns:

    df.drop(
        columns=["ingredients_list"],
        inplace=True
    )
```

```python
df.to_csv(
    "clean_recipe_dataset.csv",
    index=False,
    encoding="utf-8-sig"
)

print(
    "Saved clean_recipe_dataset.csv"
)
```
