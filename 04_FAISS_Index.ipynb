```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    "clean_recipe_dataset.csv"
)

embeddings = np.load(
    "recipe_embeddings.npy"
)

print(df.shape)
print(embeddings.shape)
```

```python
#chuẩn hóa vector
import faiss

faiss.normalize_L2(
    embeddings
)
```

```python

#tạo index
dimension = embeddings.shape[1]
index = faiss.IndexFlatIP(
    dimension
)

index.add(
    embeddings
)

print(
    index.ntotal
)
#lưu
faiss.write_index(
    index,
    "recipe.index"
)

print(
    "Saved recipe.index"
)
```

```python
#hàm tìm kiếm
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "keepitreal/vietnamese-sbert"
)

def search_recipe(
    query,
    top_k=10
):

    query_embedding = model.encode(
        [query],
        convert_to_numpy=True
    )

    faiss.normalize_L2(
        query_embedding
    )

    D, I = index.search(
        query_embedding,
        top_k
    )

    results = df.iloc[
        I[0]
    ][
        [
            "title",
            "ingredients",
            "url"
        ]
    ].copy()

    results["score"] = D[0]

    return results
```

```python
#test
search_recipe(
    "món chay ít calo"
)
```

```python
search_recipe(
    "tôi có tôm và bông cải"
)
```
