```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

import pandas as pd
import re
import time
from tqdm import tqdm
```

```python
def create_driver():

    options = Options()

    options.add_argument("--headless=new")

    options.add_argument("--disable-gpu")

    options.add_argument("--window-size=1920,1080")

    options.add_argument(
        "--disable-blink-features=AutomationControlled"
    )

    driver = webdriver.Chrome(
        options=options
    )

    return driver
```

```python
protein_keywords = [
    "gà","bò","heo","vịt",
    "cá","tôm","mực","cua",
    "trứng","đậu hũ"
]

carb_keywords = [
    "cơm","bún","phở",
    "mì","miến","xôi",
    "bánh mì"
]

vegetable_keywords = [
    "salad",
    "rau",
    "cải",
    "bí đỏ",
    "khoai tây",
    "cà chua"
]

cooking_keywords = [
    "chiên",
    "xào",
    "kho",
    "luộc",
    "hấp",
    "nướng",
    "canh",
    "cháo",
    "lẩu",
    "gỏi"
]

dessert_keywords = [
    "chè",
    "cookie",
    "cupcake",
    "cheesecake",
    "bánh flan",
    "pudding"
]

drink_keywords = [
    "trà sữa",
    "cà phê",
    "sinh tố",
    "nước ép",
    "matcha"
]

healthy_keywords = [
    "healthy",
    "eat clean",
    "giảm cân",
    "keto",
    "low carb"
]

meal_keywords = [
    "ăn sáng",
    "ăn trưa",
    "ăn tối",
    "ăn vặt",
    "nấu ăn hàng ngày"
]

keywords = list(
    set(
        protein_keywords
        + carb_keywords
        + vegetable_keywords
        + cooking_keywords
        + dessert_keywords
        + drink_keywords
        + healthy_keywords
        + meal_keywords
    )
)

print(
    "Total Keywords:",
    len(keywords)
)
```

```python
def clean_text(text):

    if not text:
        return ""

    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text.strip()
```

```python
def get_recipe_urls(keyword):

    driver = create_driver()

    urls = set()

    try:

        search_url = (
            f"https://cookpad.com/vn/tim-kiem/{keyword}"
        )

        driver.get(search_url)

        time.sleep(4)

        for _ in range(15):

            driver.execute_script(
                "window.scrollTo(0, document.body.scrollHeight)"
            )

            time.sleep(1.5)

        soup = BeautifulSoup(
            driver.page_source,
            "html.parser"
        )

        for a in soup.find_all("a"):

            href = a.get("href")

            if (
                href
                and "/cong-thuc/" in href
            ):

                if href.startswith("/"):

                    href = (
                        "https://cookpad.com"
                        + href
                    )

                href = href.split("?")[0]

                urls.add(href)

    finally:

        driver.quit()

    return list(urls)
```

```python
#craw
all_urls = []

for keyword in tqdm(keywords):

    try:

        urls = get_recipe_urls(
            keyword
        )

        print(
            keyword,
            len(urls)
        )

        all_urls.extend(urls)

    except Exception as e:

        print(
            keyword,
            e
        )

all_urls = list(
    set(all_urls)
)

print(
    "Total URLs:",
    len(all_urls)
)
```

```python
#lưu vào csv
url_df = pd.DataFrame({
    "url": all_urls
})

url_df.to_csv(
    "recipe_urls.csv",
    index=False,
    encoding="utf-8-sig"
)

print(
    "Saved recipe_urls.csv"
)
```

```python
headers = {

    "User-Agent":
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
def crawl_recipe(url):

    try:

        response = requests.get(
            url,
            headers=headers,
            timeout=20
        )

        if response.status_code != 200:

            return None

        soup = BeautifulSoup(
            response.text,
            "html.parser"
        )

        recipe_data = None

        for script in soup.find_all(
            "script",
            type="application/ld+json"
        ):

            try:

                data = json.loads(
                    script.string
                )

                if (
                    isinstance(data, dict)
                    and data.get("@type")
                    == "Recipe"
                ):

                    recipe_data = data
                    break

            except:

                pass

        if recipe_data is None:

            return None

        # -------------------
        # Ingredients
        # -------------------

        ingredients_list = recipe_data.get(
            "recipeIngredient",
            []
        )

        ingredients_text = " | ".join(
            ingredients_list
        )

        # -------------------
        # Instructions
        # -------------------

        instructions_list = []

        for step in recipe_data.get(
            "recipeInstructions",
            []
        ):

            if isinstance(
                step,
                dict
            ):

                instructions_list.append(
                    step.get(
                        "text",
                        ""
                    )
                )

        instructions_text = " | ".join(
            instructions_list
        )

        # -------------------
        # Statistics
        # -------------------

        likes = 0

        bookmarks = 0

        for item in recipe_data.get(
            "interactionStatistic",
            []
        ):

            action = item.get(
                "interactionType",
                ""
            )

            count = item.get(
                "userInteractionCount",
                0
            )

            if "LikeAction" in action:

                likes = count

            elif "BookmarkAction" in action:

                bookmarks = count

        comment_count = recipe_data.get(
            "commentCount",
            0
        )

        popularity_score = (

            likes

            + bookmarks * 2

            + comment_count * 3
        )

        cook_steps = len(
            instructions_list
        )

        # -------------------
        # Return
        # -------------------

        return {

            "title":
            recipe_data.get(
                "name",
                ""
            ),

            "description":
            recipe_data.get(
                "description",
                ""
            ),

            "ingredients":
            ingredients_text,

            "ingredients_list":
            ingredients_list,

            "instructions":
            instructions_text,

            "keywords":
            recipe_data.get(
                "keywords",
                ""
            ),

            "cuisine":
            recipe_data.get(
                "recipeCuisine",
                ""
            ),

            "author":
            recipe_data.get(
                "author",
                {}
            ).get(
                "name",
                ""
            ),

            "image_url":
            recipe_data.get(
                "image",
                ""
            ),

            "likes":
            likes,

            "bookmarks":
            bookmarks,

            "comment_count":
            comment_count,

            "popularity_score":
            popularity_score,

            "cook_steps":
            cook_steps,

            "url":
            url
        }

    except Exception as e:

        print(
            "Error:",
            url,
            e
        )

        return None

```
